```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
###########绘制热力图#########
#热力图   用颜色的差异描述数据的差异
data = np.random.rand(3,3)
print data
sns.heatmap(data)
```

```python
#指定调色板的 上限和下限 
#>0.3为一个颜色   <0.2为同一个颜色
sns.heatmap(data= data, vmin=0.2, vmax = 0.3)
```

```python
#以0为中心值  大于0 一种颜色变化  小于0一种颜色变化  并且两种颜色差异较大
data = np.random.randn(3,3)
sns.heatmap(data, center=0)
```

```python
#读取航班数据
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
#把flights表格数据通过pivot转换为矩阵图
#横轴表示 year 纵轴表示month  对应数值表示passenger数量
print flights
sns.heatmap(flights)
#通过热力图可以很清晰看出 随着时间增长  数量越来越多  
```

```python
#annot设置为True表示可以在heatmap显示数值  对应矩阵中的passenger数量
#fmt = 'd'设置字体形式  默认格式会出现数字重叠
sns.heatmap(flights, annot=True,fmt = 'd')
```

```python
sns.heatmap(flights, linewidths=0.5)
#linewidth 即设置格子之间的间距
```

```python
#设置  color bar  颜色板
sns.heatmap(flights, cmap = 'YlGnBu')
# help(sns.heatmap)
```
